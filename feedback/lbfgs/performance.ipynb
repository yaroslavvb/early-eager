````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
name = "short"
sizes = np.loadtxt("data/"+name+"_batch.csv")
eager_loss = np.loadtxt("data/"+name+"_eager_loss.csv")
pytorch_loss = np.loadtxt("data/"+name+"_pytorch_loss.csv")
eager_time = np.loadtxt("data/"+name+"_eager_time.csv")
pytorch_time = np.loadtxt("data/"+name+"_pytorch_time.csv")

plt.figure()
plt.plot(sizes, eager_loss, "r-.", label="eager")
plt.plot(sizes, pytorch_loss, "b.", label="pytorch", )
plt.xlabel('batch')
plt.ylabel('loss')
plt.title("Loss comparison")
plt.legend()
plt.show()

plt.figure()
plt.plot(sizes, eager_time, "r-.", label="eager")
plt.plot(sizes, pytorch_time, label="pytorch", )
plt.xlabel('batch')
plt.ylabel('seconds')
plt.title("Time comparison")
plt.legend()
plt.show()
````
